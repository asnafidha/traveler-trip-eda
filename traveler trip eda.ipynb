```python
#Let's keep our cost features more consistent
#cc["Transportation cost"] = cc["Transportation cost"].replace("[^\d]+", "", regex=True)
#cc["Accommodation cost"] = cc["Accommodation cost"].replace("[^\d]+", "", regex=True)

#Turn them integers
#cc["Transportation cost"] = cc["Transportation cost"].astype(int)
#cc["Accommodation cost"] = cc["Accommodation cost"].astype(int)#

import pandas as pd

# Assuming cc is your DataFrame
cc["Transportation cost"] = cc["Transportation cost"].replace("[^\d]+", "", regex=True)
cc["Accommodation cost"] = cc["Accommodation cost"].replace("[^\d]+", "", regex=True)

# Use .loc to modify the original DataFrame
cc.loc[:, "Transportation cost"] = cc["Transportation cost"].astype(int)
cc.loc[:, "Accommodation cost"] = cc["Accommodation cost"].astype(int)
```
